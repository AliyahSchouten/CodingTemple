### Problem 2:

I want to build the best Pokemon team I can and I need your help to do so! Using the API, build me a team of 6 pokemon who's base-stats are above 60. Return any additional information you think is needed to help understand why you chose that pokemon for the team!

```python

import requests

def get_pokemon(pokemon_name):
    return requests.get(f'https://pokeapi.co/api/v2/pokemon/{pokemon_name.lower()}').json()

def pokemon_info(pokemon_data):
    name = pokemon_data['name']
    abilities = [ability['ability']['name'] for ability in pokemon_data['abilities']]
    base_experience = pokemon_data['base_experience']
    sprite_url = pokemon_data['sprites']['front_shiny']
    attack = pokemon_data['stats'][1]['base_stat']
    hp = pokemon_data['stats'][0]['base_stat']
    defense = pokemon_data['stats'][2]['base_stat']

    
    return {
        'name': name,
        'abilities': abilities,
        'base_experience': base_experience,
        'sprite_url': sprite_url,
        'attack': attack,
        'hp': hp,
        'defense': defense
    }

def pokemon_team():
    best_team = []
    for x in range(1, 1001):  
        pokemon_data = get_pokemon(str(x))
        info = pokemon_info(pokemon_data)
        if info['attack'] > 60 and info['hp'] > 60 and info['defense'] > 60:
            best_team.append(info)

        if len(best_team) == 6:
            break

    return best_team

best_team = pokemon_team()

print("Best Pokemon Team:")
for pokemon in best_team:
    print(f"Name: {pokemon['name']}")
    print(f"Abilities: {', '.join(pokemon['abilities'])}")
    print(f"Base Experience: {pokemon['base_experience']}")
    print(f"Attack: {pokemon['attack']}")
    print(f"HP: {pokemon['hp']}")
    print(f"Defene: {pokemon['defense']}")
    print(f"Sprite URL: {pokemon['sprite_url']}")
    print("__________________________________________")

        

    


```

### Problem 3:

Create a function that does the following:
* Input: pokemon_name(user_defined) -> str
* Output: list object containing a dictionary of values from the response object.
* Please include the type, the most current raritey level, at least one held item, and all stats

```python
import requests

def pokemon_info(pokemon_name):
    def get_pokemon(pokemon_name):
        return requests.get(f'https://pokeapi.co/api/v2/pokemon/{pokemon_name.lower()}').json()

    
    pokemon_data = get_pokemon(pokemon_name)

    name = pokemon_data['name']
    types = [t['type']['name'] for t in pokemon_data['types']]
    rarity = pokemon_data['sprites']['versions']['generation-v']['black-white']['animated']['front_default']
    held_items = [item['item']['name'] for item in pokemon_data['held_items']]
    stats = {stat['stat']['name']: stat['base_stat'] for stat in pokemon_data['stats']}

    pokemon_info = {
        'name': name,
        'types': types,
        'rarity': rarity,
        'held_items': held_items,
        'stats': stats
    }

    return pokemon_info


while True:
    pokemon_name = input('Enter Pokemon or type "quite":').lower()
    if pokemon_name=='quite':
        break

    pokemon_info = pokemon_info(pokemon_name)

    print("Pokemon Info:")
    print(f"Name: {pokemon_info['name']}")
    print(f"Types: {', '.join(pokemon_info['types'])}")
    print(f"Rarity: {pokemon_info['rarity']}")
    print(f"Held Items: {', '.join(pokemon_info['held_items'])}")
    print("Stats:")
    for stat_name, stat_value in pokemon_info['stats'].items():
        print(f"{stat_name.capitalize()}: {stat_value}")
```
